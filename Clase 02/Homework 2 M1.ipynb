1) Abrir el archivo "Emisiones_CO2.csv" y cargar sus datos en un diccionario.

```python
dicc_emisiones = {  'cod_pais' : [],
                    'nom_pais' : [],
                    'region' : [],
                    'anio' : [],
                    'co2' : [],
                    'co2_percapita' : []}
```

```python
import os
archivo = open('Emisiones_CO2.csv', 'r')
next(archivo)
for lines in archivo:
    lines = lines.strip()
    columnas = lines.split('|')
    
    dicc_emisiones['cod_pais'].append(columnas[0])
    dicc_emisiones['nom_pais'].append(columnas[1])
    dicc_emisiones['region'].append(columnas[2])
    dicc_emisiones['anio'].append(columnas[3])
    dicc_emisiones['co2'].append(columnas[4])
    dicc_emisiones['co2_percapita'].append(columnas[5])
archivo.close()
```

```python
dicc_emisiones['cod_pais'].clear()
dicc_emisiones['nom_pais'].clear()
dicc_emisiones['region'].clear()
dicc_emisiones['anio'].clear()
dicc_emisiones['co2'].clear()
dicc_emisiones['co2_percapita'].clear()
```

e) ¿Cuál es el total de emisiones de CO2 para 'América Latina y Caribe' en el año 2010?

```python
for i, dato in enumerate(dicc_emisiones['co2']):
    dato = dato.replace('.','')
    dato = dato.replace(',','.')
    if (dato != ''):
        dato = float(dato)
    else:
        dato = None
    dicc_emisiones['co2'][i] = dato
```

```python
filtro1 = 'América Latina y Caribe'
filtro2 = '2010'
total_emisiones_co2 = 0
for i,dato in enumerate(dicc_emisiones['co2']):
    if (dicc_emisiones['region'][i] == filtro1):
        if (dicc_emisiones['anio'][i] == filtro2):
            if (dicc_emisiones['co2'][i] != None):
                total_emisiones_co2 += dicc_emisiones['co2'][i]
print(f'El total de emisiones para {filtro1} en {filtro2} es de {total_emisiones_co2} toneladas de Co2')
```
